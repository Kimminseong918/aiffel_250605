```python
import random

#Q1
class Character:
    def __init__(self, name, level, health, attack, defense):
        self.name=name
        self.level=level
        self.health=health
        self.attack=attack
        self.defense=defense
        
    #현재 체력
    def is_alive(self):
        return self.health>0
    #공격 받았을 때 
    def take_damage(self, damage):
        #현재 데미지: r_damege, 받은 데미지: damege, 본인의 방어력: self.defense
        r_damage=damage-self.defense
        if r_damage>0: #현재 데미지 0보다 크면 > 채력 감소
            self.health-=r_damage
            if self.health<0:
                self.health=0
    #공격 할 때
    def attack_target(self, target):
        damage=random.randint(1, self.attack) #데미지는 1부터 공격력 사이의 랜덤한 정수
        target.take_damage(damage) #타켓에서 데미지 입힘
        print(f"{self.name}이 {target.name}에게 {damage}만큼 공격했다...!")
        
```

```python
#Q2
class Player(Character): #Character 상속받기
    def __init__(self, name):
        super().__init__(name, 1, 100, 25, 5) #레벨 1, 체력 100, 공격력 25, 방어력 5 초기화
        self.experience=0 #경험치 속성 생성
        
    #경험치 메소드
    def gain_experience(self, exp_amount): #exp_amount: 경험치 총 양
        self.experience+=exp_amount
        self.levelup()
    def levelup(self):
        while self.experience >=50: #exp_amount가 50을 넘는 경우가 여러번 > while
            self.experience-=50
            self.level+=1
            self.attack+=10
            self.defense+=5
            
class Monster(Character): 
    def __init__(self, name, level):
    #name,level > 외부에서 전달, health, attack, defense > 함수 안에서 랜덤 생성
        health=(random.randint(10, 30))*level
        attack=(random.randint(5, 20))*level
        defense=(random.randint(1, 5))*level 
        super().__init__(name, level, health, attack, defense)
        #name, level, health, attack, defense 저장할 때 생성 후에 저장 > 마지막 작성
        #self.health=(random.randint(10, 30))*self.level > super().__init를 호출하기 전에는 self.~ 속성은 없다
```

```python
#Q3
def battle(player, monster):
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if not monster.is_alive():
            break
        monster.attack_target(player)
    if player.is_alive():
        exp_gain=monster.level*20
        player.gain_experience(exp_gain)
        print("전투 승리!")
    else:
        print("전투 패배..")
```

```python
#Q4
def main():
    monster_dict={'슬라임': 1,'고블린': 2, '오크': 3}
    
    player_name=input('플레이어 이름: ')
    player=Player(player_name)
    
    for monster_name, monster_level in monster_dict.items():
        monster=Monster(monster_name, monster_level)
        
        battle(player, monster)
        if not player.is_alive():
            print('게임오버')
            return
```

```python
if __name__ == "__main__":
    main()
```
